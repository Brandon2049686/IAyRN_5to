```python
#IMPORTAMOS LIBRERIAS
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

```python
#CARGAMOS DATOS
breast_cancer = load_breast_cancer ()
x = breast_cancer.data
y = breast_cancer.target
print("Target Names =", breast_cancer.target_names)
```

```python
#PREPROCESAMOS LOS DATOS
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#SELECCIONAMOS Y ENTRENAMOS AL MODELO
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
DecisionTreeClassifier()
```

```python
#EXACTITUD DEL MODELO
train_accuracy = clf.score(X_train, y_train)
print(f"Exactitud del modelo: {train_accuracy}")
```

```python
#PRUEBA DEL MODELO
y_pred = clf.predict(X_test)
```

```python
#PREDICCIÃ“N 1
print("Prediction:", y_pred[1])
print("Test:      ", y_test[1])
```

```python
#PREDICCIÃ“N 2
print("Prediction:", y_pred[20])
print("Test:      ", y_test[20])
```

```python
#PREDICCIÃ“N 3
print("Prediction:", y_pred[-1])
print("Test:      ", y_test[-1])
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=breast_cancer.target_names))
```

```python
#ARBOL DE DECISIÃ“N
plt.figure(figsize=(20, 15))
tree.plot_tree(clf, filled=True, feature_names=breast_cancer.feature_names, class_names=breast_cancer.target_names)
plt.title("Decision Tree for Breast Cancer")
plt.savefig("decision_tree.png", dpi=300)
plt.show()
```
